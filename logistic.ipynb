```python
#課題1
def plotmarubatu():
  list_data_y0=[[],[]]
  list_data_y1=[[],[]]
  for i in range(len(data_logT[0])):
    if data_logT[1][i] == 0:
      list_data_y0[0].append(data_logT[0][i])
      list_data_y0[1].append(data_logT[1][i])
    elif data_logT[1][i] == 1:
      list_data_y1[0].append(data_logT[0][i])
      list_data_y1[1].append(data_logT[1][i])
  
  print(len(list_data_y0[0]),len(list_data_y0[1]),list_data_y0)
  print(len(list_data_y1[0]),len(list_data_y1[1]),list_data_y1)
  plt.plot(list_data_y0[0],list_data_y0[1],'x')
  plt.plot(list_data_y1[0],list_data_y1[1],'o')
  plt.show()
  return
```

```python
def func1(w0,w1):
  S = 0
  for i in range(len(data_logT[1])):
    S += -data_logT[1][i]*w0-data_logT[1][i]*w1*data_logT[0][i]+np.log(1+np.e**w0*np.e**(w1*data_logT[0][i]))
  return S
def func1_dw0(w0,w1):
  Sw0 = 0
  for i in range(len(data_logT[1])):
    Sw0 += (np.e**w0*np.e**(w1*data_logT[0][i]))/(1+np.e**w0*np.e**(w1*data_logT[0][i])) - data_logT[1][i]
  return Sw0
def func1_dw1(w0,w1):
  Sw1 = 0
  for i in range(len(data_logT[1])):
    Sw1 += (data_logT[0][i]*np.e**w0*np.e**(w1*data_logT[0][i]))/(1+np.e**w0*np.e**(w1*data_logT[0][i])) - data_logT[0][i]*data_logT[1][i]
  return Sw1
```

```python
#課題3
def grad_method(f,f_dw0,f_dw1,eps,step,inti_w0,inti_w1,MAXLOOP):
  data_w0 = []
  data_w1 = []
  data_dw0 = []
  data_dw1 = []

  w0 = inti_w0
  w1 = inti_w1
  for i in range(MAXLOOP):
    if not -eps <= f_dw0(w0,w1) <= eps or not -eps <= f_dw1(w0,w1) <= eps:
      w0 = w0 - step*f_dw0(w0,w1)
      w1 = w1 - step*f_dw1(w0,w1)
      data_w0.append(w0)
      data_w1.append(w1)
      data_dw0.append(f_dw0(w0,w1))
      data_dw1.append(f_dw1(w0,w1))
    else:
      data_w0.append(w0)
      data_w1.append(w1)
      data_dw0.append(f_dw0(w0,w1))
      data_dw1.append(f_dw1(w0,w1))
      break
      
  print(data_w0)
  print(data_w1)
  print(data_dw0)
  print(data_dw1)

  W0 = np.arange(0,len(data_dw0),1)
  plt.plot(W0,data_dw0,'o')
  plt.show()
  W1 = np.arange(0,len(data_dw1),1)
  plt.plot(W1,data_dw1,'o')
  plt.show()
  return w0,w1,f(w0,w1),f_dw0(w0,w1),f_dw1(w0,w1),i
```

```python
grad_method(func1,func1_dw0,func1_dw1,0.0001,0.5,5,-3,1000)
```

```python
#課題4
def log_reg(f,w0,w1):
  list_data_y0=[[],[]]
  list_data_y1=[[],[]]
  for i in range(len(data_logT[0])):
    if data_logT[1][i] == 0:
      list_data_y0[0].append(data_logT[0][i])
      list_data_y0[1].append(data_logT[1][i])
    elif data_logT[1][i] == 1:
      list_data_y1[0].append(data_logT[0][i])
      list_data_y1[1].append(data_logT[1][i])
  
  print(len(list_data_y0[0]),len(list_data_y0[1]),list_data_y0)
  print(len(list_data_y1[0]),len(list_data_y1[1]),list_data_y1)
  plt.plot(list_data_y0[0],list_data_y0[1],'x')
  plt.plot(list_data_y1[0],list_data_y1[1],'o')
  X = np.arange(-7,7,0.1)
  Y = f(X,w0,w1)
  plt.plot(X,Y)
  plt.show()
```

```python
def Prob(x,w0,w1):
  return 1-1/(1+np.e**w0*np.e**(w1*x))
```

```python
#課題5
def logistic_test(f,w0,w1,alpha):
  test_y = []
  bunsi = 0
  for i in range(len(data_testT[0])):
    if f(data_testT[0][i],w0,w1) >= alpha:
      test_y.append(1)
    else:
      test_y.append(0)
  print(test_y)
  for j in range(len(data_testT[1])):
    if test_y[j] == data_testT[1][j]:
      bunsi += 1
    else:
      bunsi += 0
  print(bunsi)
  return print('{}個のうち、{}個が合っていたので、整合率は{}'.format(len(data_testT[0]),bunsi,bunsi/len(data_testT[0])))
```

```python
logistic_test(Prob,5.905558156492452,-3.384429048889434,0.5)
```
